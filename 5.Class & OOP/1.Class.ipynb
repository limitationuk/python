```python
#Class는 객체의 단위(Member variable(field) + Member function(method))

#Self는 메모리에 올려진 자기 자신의 instance를 가리킴
#Self를 통해 접근하는 값은 instance 별로 서로 다름
#일반적으로 method들은 첫 인자로 self를 가짐
#앞에 붙은 인스턴스 객체를 가리키는 참조

class Car :
    speed = 0
    color = "" # Public
    _weigh = 0 # Public, protected conventionaly
    __price = 0 # Private
    
    def accel(self):
        self.speed += 10
    
    def brake(self):
        if self.speed >= 0:
            self.speed -=5
    
    def set_color(self, color):
        self.color = color

    def print_info(self):
        print(
            f"A {self.__price} value",
            f"{self.color} car",
            f"speed {self.speed} and weight {self._weigh}")

#Instance는 class로 부터 생성한 객체

red_car = Car()
blue_car = Car()
yellow_car = Car()

red_car.set_color("red")
blue_car.set_color("blue")
yellow_car.set_color("yellow")

red_car.accel()
blue_car.accel()
yellow_car.accel()
yellow_car.accel()
blue_car.brake()

red_car.print_info()
blue_car.print_info()
yellow_car.print_info()

#Python에서는 느슨한 접근제어를 제공
#멤버 변수/함수 이름이 "_"로 시작되면 외부에서 접근 불가
#_Class_<멤버이름>으로 mangle됨

car = Car()
# 동작
car.color = "purple" 
car._weigh = "1.5t" 
# 동작 안함
car.__price = "100" 
# 맹글된 이름으로 호출하면 동작하나 비추
car._Car__price = 10000
# ex)car._Car__Private_method(100)
car.print_info()
```

```python
#Class생성자(constructor)는 instance 생성 초기에 호출됨
#__init__(self) 형식으로 작성, 인자를 받을 수 있음
class Car:
    def __init__(self, color, weight, price):
        self.speed = 0
        self.color = color
        self._weight = weight
        self.__price = price
    
    def print_info(self):
        print(
            f"A {self.__price} value",
            f"{self.color} car",
            f"speed {self.speed} and weight {self._weight}")

car = Car("purple", "1.5t", 10000)
car.print_info()        
```

```python
#Class Member 변수는 instance 간에 공유됨
#self.<Variable_name>: 인스턴스 변수
#Car.<Variable_name>: 클래스 멤버 변수 (객체간에 공유됨)
class Car:
    count = 0

    def __init__ (self, color, weight, price):
        self.speed = 0
        self.color = color
        self._weight = weight
        self.__price = price
        # Class member variable
        Car.count += 1
    
    #Class Member 함수
    #@classmethod 어노테이션으로 정의
    #Parameter로 class가 넘겨짐
    #e.g Factory method:
    @classmethod
    def create_red_car(cls, weight, price): #첫 번째 자리가 self가 아닌 class 자체
        return cls("red", weight, price)
    
    @classmethod
    def create_blue_car(cls, weight, price): #첫 번째 자리가 self가 아닌 class 자체
        return cls("blue", weight, price)
    
    @classmethod
    def create_yellow_car(cls, weight, price): #첫 번째 자리가 self가 아닌 class 자체
        return cls("yellow", weight, price)

    # Static Memter 함수
    # Class member 함수 처럼 instance 없이도 호출 가능
    # Class를 인자로 받지 않음
    # e.g Km/h를 mph로 변환
    @staticmethod
    def kmh_to_mph (kmh_value) :
         return kmh_value * 0.621371   

    def print_info(self):
            print(
                f"Count: {Car.count}.",
                f"A {self.__price} value",
                f"{self.color} car",
                f"speed {self.speed} and weight {self._weight}")
    
# Create instance
cars = [Car("red", "1.5t", 1000) for _ in range(10)]
cars[0].print_info()
cars[-1].print_info()

# Create instance
re = Car.create_red_car("1.5t", 100)
re.print_info()

bc = Car.create_blue_car("1.5t", 1000)
bc.print_info()

yc = Car.create_yellow_car("1.5", 10000)
yc.print_info()

# Create instance
print(f"100km/h -> {Car.kmh_to_mph(100):.2f} Mph") # 소수점 자리수 표시 :.2f
```
